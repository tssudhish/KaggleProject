# Part D

Using normalized data same as Part B and using number of epoch as 50. But modifying NN model with two more hidden layers of 10 nodes.

### Import Required Packages

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

### Normalized Data

```python
X_norm = (X - X.mean()) / X.std()
X_norm.head()
```

```python
# define regression model
def regression_model(n_cols):
    # create model
    model = Sequential()
    # hidden layer of 10 nodes and 'relu' activation function.
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))   
    # output layer
    model.add(Dense(1))
    
    # compile model using 'adam' optimizer
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

### Make the initial model


```python
number_of_epoch=50
split_amount=0.3
# Using Normalized data
X_train, X_test, y_train, y_test = train_test_split( X_norm, y, test_size=split_amount, random_state=4)
n_cols = X_train.shape[1] # number of features
# build the model
model = regression_model(n_cols)
# fit the model
# Note only train data of the model is used to create the model. 
# validation split is assumed as 0 to use all the X_train data into model.
# epoch verbose is switched on.
# Increase number of epoch to 100.

model.fit(X_train, y_train, validation_split=0.0, epochs=number_of_epoch, verbose=2)

# predict for test data
y_predict=model.predict(X_test.values).flatten()

# get mean squared error value for the predicted values with that of test data
mse=mean_squared_error(y_test.values, y_predict)

print("Mean Squared Values of the created model: {:4f}".format(mse))
```

```python
def create_model_and_report_mse(X,y,number_of_iterations, number_of_epoch):
    mse_values=list()
    for i in range(number_of_iterations):
        print("Working on Iteration: {}".format(i))
        ## Split data into train and test
        #### Keeping 30% of data for testing
        split_amount=0.3
        # Create Train and Test data
        X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=split_amount, random_state=4)

        n_cols = X_train.shape[1] # number of features

        # build the model
        model = regression_model(n_cols)

        # fit the model
        # Note only train data of the model is used to create the model. 
        # validation split is assumed as 0 to use all the data into model.
        # epoch verbose is switched on.
        model.fit(X_train, y_train, validation_split=0.0, epochs=number_of_epoch, verbose=2)

        # predict for test data
        y_predict=model.predict(X_test.values).flatten()

        # get mean squared error value for the predicted values with that of test data
        mse=mean_squared_error(y_test.values, y_predict)

        # append to the list
        mse_values.append(mse)
    return mse_values

```

```python
#----------------------------------------------------------
# Please note X,y is split into train and test inside the definition hence the whole data is sent in.
# Using Normalized data
number_of_iterations=50
number_of_epoch=50

mse_values=create_model_and_report_mse(X_norm,y,number_of_iterations,number_of_epoch)
print(mse_values)
```
